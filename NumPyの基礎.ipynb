NumPyの配列を生成

```python
import numpy as np

a = np.array([0, 1, 2, 3, 4, 5]) #PythonのリストからNumPyの配列を作る
print(a)
```

```python
import numpy as np

b = np.array([[0, 1, 2], [3, 4, 5]]) #2重のリストからNumPyの2次元配列を作る
print(b)
```

```python
import numpy as np

c = np.array([[[0, 1, 2], [3, 4, 5]], [[5, 4, 3], [2, 1, 0]]]) #3重のリストからNumPyの3次元配列を作る
print(c)
```

```python
import numpy as np

d = np.zeros(8) #0が8個格納された配列
print(d)

e = np.ones(8) #1が8個格納された配列
print(e)

f = np.arange(8) #0から7までが格納された配列
print(f)
```

配列の演算

```python
import numpy as np

a = np.array([[0, 1, 2], [3, 4, 5]]) #2次元配列

print(a)
print()
print(a + 3) #各要素に3を足す
print()
print(a * 3) #各要素に3を掛ける
```

```python
b = np.array([[0, 1, 2], [3, 4, 5]]) #2次元配列

c = np.array([[2, 0, 1], [5, 3, 4]]) #2次元配列

print(b)
print()
print(c)
print()
print(b + c)
print()
print(b * c)
```

要素へのアクセス

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])
print(a[3]) #インデックスを指定
```

```python
a[2] = 9
print(a)
```

```python
b = np.array([[0, 1, 2], [3, 4, 5]])

print(b[1, 2]) #b[1][2]と同じ
```

```python
b[1, 2] = 9

print(b)
```

```python
c = np.array([[0, 1, 2], [3, 4, 5]])

print(c[1, :]) #インデックスが1の行を取得

print()

c[:, 1] = np.array([6, 7]) #インデックスが1の列を入れ替え
print(c)
```

関数と配列

```python
import numpy as np

def my_func(x):
  y = x * 2 + 1
  return y

a = np.array([[0, 1, 2], [3, 4, 5]]) #2次元配列
b = my_func(a) #引数として配列を渡し、返り値として配列を受け取る

print(b)
```

NumPyの様々な機能

```python
import numpy as np

a = np.array([[0, 1, 2], [3, 4, 5]]) #2次元配列

print("合計:", np.sum(a))
print("平均:", np.average(a))
print("最大値:", np.max(a))
print("最小値:", np.min(a))
```

演習

```python
import numpy as np

a = np.array([[0, 1, 2], [3, 4, 5]])
b = np.array([[5, 4, 3], [2, 1, 0]])

print(a + b) #和
print()
print(a - b) #差
print()
print(a * b) #積
```
